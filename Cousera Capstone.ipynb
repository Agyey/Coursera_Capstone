```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut
import geopy

def do_geocode(address, attempt=1, max_attempts=5):
    try:
        geolocator = Nominatim(user_agent="foursquare_agent")
        location = geolocator.geocode(address)
        return location
    except GeocoderTimedOut:
        if attempt <= max_attempts:
            return do_geocode(address, attempt=attempt+1)
        return None
```

```python
neighbourhoods = {}
```

```python
URL = 'https://en.wikipedia.org/wiki/List_of_neighbourhoods_in_Toronto'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
tds = soup.find_all('td', class_=['navbox-list navbox-odd hlist','navbox-list navbox-even hlist'])
trs = soup.find_all('tr')
not_found = []
for tr in trs:
    th = tr.find_all('th')
    td = tr.find_all('td', class_=['navbox-list navbox-odd hlist','navbox-list navbox-even hlist'])
    if td:
        th = th[0]
        td = td[0]
        links = td.find_all('a')
        print("District : ",th.text)
        for link in links:
            print(link.text, link['href'])
            if neighbourhoods and link.text in neighbourhoods["Neighbourhood"]:
                continue
            URL = 'https://en.wikipedia.org' + link['href']
            page = requests.get(URL)
            soup = BeautifulSoup(page.content, 'html.parser')
            spans = soup.find_all('span', class_=['geo-default'])
            if spans:
                text = spans[0].text
                if '/' in text:
                    text = text.split("/")[0]
                lat, long = text.strip().split(" ")
                if "°" in lat:
                    lat, rem = lat.split("°")
                    if "′" in rem:
                        rem, misc = rem.split("′")
                        lat = float(lat) + float(rem)/60
                    else:
                        lat = float(lat)
                if "°" in long:
                    long, rem = long.split("°")
                    if "′" in rem:
                        rem, misc = rem.split("′")
                        long = float(long) + float(rem)/60
                    else:
                        long = float(long)
                print("Longitude : ", -long,"Latitude : ",lat)
                longitude = -long
                latitude = lat
            else:
                address = link.text + ', ' + th.text + ', Toronto'
#                 geolocator = Nominatim(user_agent="foursquare_agent")
#                 location = geolocator.geocode(address)
                location = do_geocode(address)
                if location:
                    latitude = location.latitude
                    longitude = location.longitude
                    print("Longitude : ", longitude,"Latitude : ",latitude)
                else:
                    print("Co-ordinates Not Found ",link.text)
                    not_found.append(link.text)
                    continue   
            if neighbourhoods:
                neighbourhoods["District"].append(th.text)
                neighbourhoods["Neighbourhood"].append(link.text)
                neighbourhoods["Longitude"].append(longitude)
                neighbourhoods["Latitude"].append(latitude)
            else:
                neighbourhoods["District"] = [th.text]
                neighbourhoods["Neighbourhood"] = [link.text]
                neighbourhoods["Longitude"] = [longitude]
                neighbourhoods["Latitude"] = [latitude]
                
```

```python
neighbourhoodsDF = pd.DataFrame(neighbourhoods)
neighbourhoodsDF.to_csv("neighbourhoods.csv",index=False)
neighbourhoodsDF
```
